```python
import seaborn as sns
from matplotlib import pyplot as plt
from sqlalchemy import create_engine, text
from sqlalchemy import Numeric
import pandas as pd
from jproperties import Properties
```

```python
p = Properties()
with open('app.properties', 'rb') as f:
    
    p.load(f, 'utf-8')

uri = p.get('uri').data
```

```python
engine = create_engine(uri)

def run_sql(query):

    with engine.connect() as conn:
        query_result = conn.execute(text(query)).fetchall()

    return pd.DataFrame(query_result)
```

```python
q_count_kind = run_sql("""
                        select kind, count(*) 
                        from zap.silver_zapimoveis sz 
                        group by kind order by count(*) desc """)



q_count_kind['percent'] = round(q_count_kind['count'] / q_count_kind['count'].sum() * 100, 2)
q_count_kind = q_count_kind.set_index('kind')
q_count_kind
```

```python

q_count_kind.to_csv('./Tableau/listing_per_re_type.csv', index = False)

q_gold_apartment_avg_prices = run_sql("""select * from zap.gold_apartment_avg_prices gaap order by gaap.price_rent_per_m2_avg desc""")
q_gold_apartment_avg_prices.to_csv('./Tableau/gold_apartment_avg_prices.csv', index=False)


q_gold_flat_avg_prices = run_sql("""select * from zap.gold_flat_kit_studio_avg_prices gfap order by price_rent_per_m2_avg DESC""")
q_gold_apartment_avg_prices.to_csv('./Tableau/gold_flat_avg_prices.csv', index=False)


q_gold_floorsize_ranked = run_sql("""SELECT * FROM zap.gold_floorsize_ranked_avg_prices""")
q_gold_floorsize_ranked.to_csv('./Tableau/gold_floorsize_ranked_avg.csv', index = False)
```
